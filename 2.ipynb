Искажение видео

```python
import cv2
import numpy as np

# Открываем видео
video = cv2.VideoCapture('./data/Med_video.mp4')

while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break
    # Показываем кадр
    cv2.imshow('Video_original', frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()


# Открываем видео на обработку
video = cv2.VideoCapture('./data/Med_video.mp4')
while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break
        
    width = frame.shape[1]

    # Добавляем шум
    noise = np.random.normal(0, 0.5, frame.shape).astype(np.uint8)
    noisy_frame = cv2.add(frame, noise)

    # Искажаем цвета
    distorted_frame = cv2.cvtColor(noisy_frame, cv2.COLOR_BGR2HSV)
    distorted_frame[:, :, 0] = (distorted_frame[:, :, 0] + 90) % 180
    distorted_frame = cv2.cvtColor(distorted_frame, cv2.COLOR_HSV2BGR)

    # Три полосы
    stripe_width = 2
    for i in range(3):
        stripe_pos = np.random.randint(0, width - stripe_width)
        distorted_frame[:, stripe_pos:stripe_pos + stripe_width, :] = 0
        
    
    # Меняем яркость
    
    brightness_factor = np.random.uniform(low=0.8, high=1.3)
    bright_frame = cv2.convertScaleAbs(distorted_frame, alpha=brightness_factor)

    # Показываем обработанный кадр
    cv2.imshow('Video_new', bright_frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
