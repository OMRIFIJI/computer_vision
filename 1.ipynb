```python
import cv2
import numpy as np
import random

BLACK = (0, 0, 0)
YELLOW = (0, 255, 255)
BLUE = (255, 0, 0)

background = cv2.imread("./data/background.jpg")

height, width, _ = background.shape

backman_radius = 20
backman_x, backman_y = width // 2, height // 2

starting_screen = cv2.imread("./data/start.jpg")
cherry = cv2.imread("./data/cherry.jpg")


# Вишня
cherry_flag = 0
cherry_height, cherry_width, tmp = cherry.shape
print(cherry_height, cherry_width)

# SCORE
score = 0
font = cv2.FONT_HERSHEY_SIMPLEX
org = (10, 30)
fontScale = 1
thickness = 2

# Wall
wall_x0, wall_x1 = 201, 214
wall_y0, wall_y1 = 0, 309



cv2.imshow("Russian game: The Backman. Press ENTER.", starting_screen)

while True:
    key = cv2.waitKey(1)
    if key == 13:
        cv2.destroyAllWindows()
        break;
        
del starting_screen

def is_wall(x, y):
    if (wall_y1 + backman_radius > y > wall_y0):
        if (wall_x1 + backman_radius> x > wall_x0 - backman_radius):
            return True
    return False
    

while True:
    image = background.copy()

    key = cv2.waitKey(1)
    if key == ord('w'):
        if not is_wall(backman_x, backman_y-8):
            backman_y -= 8
    elif key == ord('a'):
        if not is_wall(backman_x-8, backman_y):
            backman_x -= 8
    elif key == ord('s'):
        if not is_wall(backman_x, backman_y+8):
            backman_y += 8
    elif key == ord('d'):
        if not is_wall(backman_x+8, backman_y):
            backman_x += 8
        
    if key == 82:  # Код клавиши вверх
        backman_y -= 8
    elif key == 84:  # Код клавиши вниз
        backman_y += 8
    elif key == 81:  # Код клавиши влево
        backman_x -= 8
    elif key == 83:  # Код клавиши вправо
        backman_x += 8

    if backman_x > width-backman_radius:
        backman_x -= 8
    elif backman_x < backman_radius:
        backman_x += 8
    elif backman_y > height-backman_radius:
        backman_y -= 8
    elif backman_y < backman_radius:
        backman_y += 8
    
    if cherry_flag == 0:
        cherry_x = random.randint(0, width - cherry_width)
        cherry_y = random.randint(0, height - cherry_height)    
        cherry_flag = 1
        
        
    # Вишня
    if abs(backman_y - (cherry_y + cherry_height // 2 ) ) + abs(backman_x - (cherry_x + cherry_width // 2) ) < 20:
        cherry_flag = 0
        score +=1;
        
    
    image[cherry_y:cherry_y+cherry_height, cherry_x:cherry_x+cherry_width, :] = cherry
    
    
    image = cv2.putText(image, 'SCORE: ' + str(score), org, font, fontScale, BLUE, thickness, cv2.LINE_AA)
        
    # Стена
    cv2.rectangle(image, (wall_x0, wall_y0),
                  (wall_x1, wall_y1), BLUE, -1)
        
        
    # Бэкмэн
    cv2.rectangle(image, (backman_x - backman_radius, backman_y - backman_radius),
                  (backman_x + backman_radius, backman_y + backman_radius), YELLOW, -1)
    cv2.circle(image, (backman_x + backman_radius // 2, backman_y - backman_radius // 2), 5, BLACK, -1)
    
    cv2.circle(image, (backman_x - backman_radius // 2, backman_y - backman_radius // 2), 5, BLACK, -1)

    cv2.imshow("Russian game: The Backman. Press escape to quit.", image)
    

    if key == 27:  
        break

cv2.destroyAllWindows()
```
